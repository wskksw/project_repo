# Inference Metrics Analysis

This notebook wraps the reusable analysis helper in `inference_eval.analysis.metrics_analysis` and displays the generated tables.

```python
from pathlib import Path
import sys

import pandas as pd
from IPython.display import display

CWD = Path.cwd().resolve()
REPO_ROOT = next((path for path in [CWD, *CWD.parents] if (path / "inference_eval").exists()), None)
if REPO_ROOT is None:
    raise RuntimeError(f"Could not locate repo root from {CWD}")
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

from inference_eval.analysis.metrics_analysis import build_analysis_outputs
```

```python
RUN_DIR = REPO_ROOT / "inference_eval" / "results"
OUTPUT_DIR = REPO_ROOT / "inference_eval" / "analysis" / "outputs"
outputs = build_analysis_outputs(RUN_DIR, OUTPUT_DIR)
outputs
```

## Summary Table

Variant-level table using the same primary metrics shown in `metrics_summary.md`.

```python
summary_df = pd.read_csv(outputs["summary_table"])
display(summary_df)
```

## Grouped By Prompt Strategy

Mean primary metrics across variants within each prompt strategy.

```python
strategy_df = pd.read_csv(outputs["prompt_strategy_comparison"])
display(strategy_df)
```

## Grouped By Family

Mean primary metrics grouped by provider/model family.

```python
family_df = pd.read_csv(outputs["family_comparison"])
display(family_df)
```

## Metric-Centric View

This flips the summary table so metrics become rows and variants become columns.

```python
metric_view_df = pd.read_csv(outputs["metric_view"])
display(metric_view_df)
```

## Primary Metric Leaderboard

Ranks variants separately for each primary metric.

```python
leaderboard_df = pd.read_csv(outputs["primary_metric_leaderboard"])
display(leaderboard_df.head(30))
```

## Coverage

Scorable-pair counts `n` for the primary metrics. This is the first place to look when a metric value looks unstable or missing.

```python
coverage_df = pd.read_csv(outputs["primary_metric_coverage"])
display(coverage_df)
```
